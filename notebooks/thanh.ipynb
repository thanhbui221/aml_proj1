Player strength and postition strength

Two columns per player:
1. player_rating = PLAYER_RATING_5_last_match_average (0-100)
2. player_contribution = sum(or average) of all strength/contribution stats (goals, assisits, tackels, etc) according to his postition.

Per (match ID, team, position), we aggregate both.

```python
RATING_COL = "PLAYER_RATING_5_last_match_average"

CONTRIBUTION_COLS = [
    "PLAYER_GOALS_season_average",
    "PLAYER_ASSISTS_season_average",
    "PLAYER_KEY_PASSES_season_average",
    "PLAYER_TACKLES_season_average",
    "PLAYER_INTERCEPTIONS_season_average",
    "PLAYER_CLEARANCES_season_average",
    "PLAYER_DUELS_WON_season_average",
    "PLAYER_SABES_season_average",
    "PLAYER_ACCURATE_PASSES_PERCENTAGE_season_average"
]

def add_player_two_score(player_df, contribution_agg="mean"):
    player_df["player_rating"] = pd.to_numeric(
        player_df[RATING_COL], errors="coerce"
    ).fillna(0) if RATING_COL in player_df.columns else 0.0

    avai = [c for c in CONTRIBUTION_COLS if c in player_df.columns]
    if avai:
        contrib = player_df[avai].apply(pd.to_numeric, errors="coerce").fillna(0)
        player_df["player_contribution"] = contrib.sum(axis=1) if contribution_agg == "sum" else contrib.mean(axis=1)

    else:
        player_df["player_contribution"] = 0.0

    return player_df 


def pos_strength_per_match(player_df, side_prefix):
    player_df = player_df.copy()
    add_player_two_score(player_df, contribution_agg="sum")
    player_df["position"] = player_df["POSITION"].astype(str).str.strip().str.lower()
    player_df = player_df[player_df["position"].str.len() > 0]

    by_pos = player_df.groupby(["ID", "position"], as_index=False).agg(
        rating_avg=("player_rating", "mean"),
        contribution_sum=("player_contribution", "sum")
    )

    out = by_pos.pivot_table(
        index="ID",
        columns="position",
        values=["rating_avg", "contribution_sum"],
        aggfunc="first"
    )

    out.columns = [f"{side_prefix}_{pos}_{m}" for m, pos in out.columns]
    out = out.reset_index()

    return out

```

```python
home_player = pd.read_csv(TRAIN_DIR / "train_home_player_statistics_df.csv")
away_player = pd.read_csv(TRAIN_DIR / "train_away_player_statistics_df.csv")

home_pos_strength = pos_strength_per_match(home_player, "home")
away_pos_strength = pos_strength_per_match(away_player, "away")

pos_strength = home_pos_strength.merge(away_pos_strength, on="ID", how="outer").fillna(0)
pos_strength.head()
```

```python
X_team = X_team.merge(pos_strength, on="ID", how="left")
pos_cols = [c for c in pos_strength.columns if c != "ID"]
X_team[pos_cols] = X_team[pos_cols].fillna(0)
X_team.head()
```

```python
# drop IDS, lavels, and text cols
id_cols = ["ID", "home_ID", "away_ID"]
label_cols = ["HOME_WINS", "DRAW", "AWAY_WINS"]
text_cols = [c for c in X_team.columns if X_team[c].dtype == object]
drop_cols = list[str](set[str](id_cols+label_cols+text_cols))

all_feature_cols = [c for c in X_team.columns if c not in drop_cols]
X = X_team[all_feature_cols]
X.head()
```

```python
def get_reduced(columns, suffixes=("_5_last_match_average", "_season_average")):
    return [c for c in columns if any(c.endswith(s) for s in suffixes)]

feature_cols = get_reduced(all_feature_cols)
pos_feature_cols = [c for c in X_team.columns if ("_rating_avg" in c or "_contribution_sum" in c)
                    and (c.startswith("home_") or c.startswith("away_"))]
feature_cols += pos_feature_cols
feature_cols
```

```python
X = X_team[feature_cols].copy()
X.describe(include="all")
```

```python
y = X_team[["HOME_WINS", "DRAW", "AWAY_WINS"]]
y_label = np.argmax(y.values, axis=1)

X.shape
```

```python
X = X.fillna(X.median())

from sklearn.model_selection import cross_val_score 
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier(n_estimators=100, max_depth=5, random_state=42)
scores = cross_val_score(model, X, y_label, cv=5, scoring="accuracy")
```

Benchmark

```python
train_median = X.median()
X_filled = X.fillna(train_median)
model.fit(X_filled, y_label)
train_pred = model.predict(X_filled)
model_train_acc = (train_pred == y_label).mean()
baseline_pred = np.zeros(len(y_label), dtype=int) # 0 -> HOME_WINS
base_train_acc = (baseline_pred == y_label).mean()
```

```python
TEST_DIR = DATA_DIR / "Test_Data" 
test_home_team = pd.read_csv(TEST_DIR / "test_home_team_statistics_df.csv")
test_away_team = pd.read_csv(TEST_DIR / "test_away_team_statistics_df.csv")
test_home_team = test_home_team.rename(columns={c: f"home_{c}" for c in test_home_team.columns if c != "ID"})
test_away_team = test_away_team.rename(columns={c: f"away_{c}" for c in test_away_team.columns if c != "ID"})
X_team_test = test_home_team.merge(test_away_team, on="ID", how="inner")

test_home_players = pd.read_csv(TEST_DIR / "test_home_player_statistics_df.csv")
test_away_players = pd.read_csv(TEST_DIR / "test_away_player_statistics_df.csv")

home_pos_test = pos_strength_per_match(test_home_players, "home")
away_pos_test = pos_strength_per_match(test_away_players, "away")
pos_strength_test = home_pos_test.merge(away_pos_test, on="ID", how="outer").fillna(0)
X_team_test = X_team_test.merge(pos_strength_test, on="ID", how="left")
pos_cols = [c for c in pos_strength_test.columns if c != "ID"]
X_team_test[pos_cols] = X_team_test[pos_cols].fillna(0)


X_test = X_team_test.reindex(columns=feature_cols).fillna(train_median)
```

```python
test_idx = X_team_test["ID"]
test_pred_label = model.predict(X_test)
sub = pd.DataFrame({"ID": test_idx})
sub["HOME_WINS"] = (test_pred_label == 0).astype(int)
sub["DRAW"] = (test_pred_label == 1).astype(int)
sub["AWAY_WINS"] = (test_pred_label == 2).astype(int)
sub.to_csv(DATA_DIR/"thanh_submission.csv", index=False)
```
